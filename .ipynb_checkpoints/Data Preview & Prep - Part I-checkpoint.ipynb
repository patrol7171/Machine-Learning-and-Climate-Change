```python
import pandas as pd
import numpy as np
import datetime
```

# Earth Surface Temperature Datasets

### Data by City - US Only (includes longitude & latitude)

```python
us_cities_GLT_df = pd.read_csv("All Datasets/climate-change-earth-surface-temperature-data/US_Cities_GLT.csv", low_memory=False)
state_GLT_df = pd.read_csv("All Datasets/climate-change-earth-surface-temperature-data/GlobalLandTemperaturesByState.csv", low_memory=False)
country_GLT_df = pd.read_csv("All Datasets/climate-change-earth-surface-temperature-data/GlobalLandTemperaturesByState.csv", low_memory=False)
GT_df = pd.read_csv("All Datasets/climate-change-earth-surface-temperature-data/GlobalTemperatures.csv", low_memory=False)
```

```python
us_cities_GLT_df.head()
```

```python
# Set date as the index and drop the extra date column
us_cities_GLT_df = us_cities_GLT_df.set_index(pd.DatetimeIndex(us_cities_GLT_df['dt']))
us_cities_GLT_df.drop('dt', axis=1, inplace=True)
us_cities_GLT_df.head()
```

```python
us_cities_GLT_df.info()
```

### Data By State - All Countries

```python
state_GLT_df.head()
```

```python
# Drop rows with NaN values and set date as index
new_state_GLT_df = state_GLT_df.dropna()

# Set date as the index and drop the extra date column
new_state_GLT_df = new_state_GLT_df.set_index(pd.DatetimeIndex(new_state_GLT_df['dt']))
new_state_GLT_df.drop('dt', axis=1, inplace=True)

# Only use the date from 1918 - 2013
state_GLT_100_df = new_state_GLT_df['1918-01-01':'2013-06-01']
state_GLT_100_df.head()
```

```python
state_GLT_100_df.info()
```

```python
state_GLT_100_df.index
```

```python
#Filter out US info
us_states_df = state_GLT_100_df.query("Country == 'United States'")
us_states_df.head()
```

### Data By Country

```python
country_GLT_df.head()
```

```python
# Drop rows with NaN values and set date as index
new_country_GLT_df = country_GLT_df.dropna()

# Set date as the index and drop the extra date column
new_country_GLT_df = new_country_GLT_df.set_index(pd.DatetimeIndex(new_country_GLT_df['dt']))
new_country_GLT_df.drop('dt', axis=1, inplace=True)

# Only use the date from 1918 - 2013
country_GLT_100_df = new_country_GLT_df['1918-01-01':'2013-06-01']
country_GLT_100_df.head()
```
